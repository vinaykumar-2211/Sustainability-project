```
# ============================================
# 1. IMPORT LIBRARIES
# ============================================
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import os
import zipfile # Import zipfile for extraction
import shutil # Import shutil for data splitting
import random # Import random for data splitting

# ============================================
# 2. GOOGLE DRIVE MOUNT (OPTIONAL)
# ============================================
from google.colab import drive
drive.mount('/content/drive')

# ============================================
# 2.5. EXTRACT DATASET (ADDED)
# ============================================
zip_path = '/content/reduced_archive_25mb.zip'
extract_path = '/content/dataset'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Unzip the archive
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print(f"Dataset extracted to {extract_path}")

# ============================================
# 2.6. SPLIT DATA INTO TRAIN/TEST (ADDED)
# ============================================
# Base directory where all class folders are located (identified from previous inspection)
nested_dir = 'Faulty_solar_panel'
base_data_dir = os.path.join(extract_path, nested_dir)

# Define the new train and test directories that will be created
new_train_dir = os.path.join(base_data_dir, 'train')
new_test_dir = os.path.join(base_data_dir, 'test')

# Define target binary classes based on original problem context (e.g., Damaged vs. Undamaged)
# Assuming 'Electrical-damage', 'Physical-Damage', 'Bird-drop', 'Snow-Covered', 'Dusty' are "damaged"
# And 'Clean' is "undamaged" for binary classification
damaged_classes = ['Electrical-damage', 'Physical-Damage', 'Bird-drop', 'Snow-Covered', 'Dusty']
undamaged_classes = ['Clean']

# Create new structure: base_data_dir/train/damaged, base_data_dir/train/undamaged, etc.
os.makedirs(os.path.join(new_train_dir, 'damaged'), exist_ok=True)
os.makedirs(os.path.join(new_train_dir, 'undamaged'), exist_ok=True)
os.makedirs(os.path.join(new_test_dir, 'damaged'), exist_ok=True)
os.makedirs(os.path.join(new_test_dir, 'undamaged'), exist_ok=True)

# Define a split ratio (e.g., 80% train, 20% test)
train_split_ratio = 0.8

print(f"Splitting data from {base_data_dir} into {new_train_dir} and {new_test_dir}...")

# Iterate through each original class folder
for class_folder_name in os.listdir(base_data_dir):
    original_class_path = os.path.join(base_data_dir, class_folder_name)

    if os.path.isdir(original_class_path):
        # Exclude the newly created 'train' and 'test' directories from being processed again
        if class_folder_name in ['train', 'test']:
            continue

        images = [f for f in os.listdir(original_class_path) if os.path.isfile(os.path.join(original_class_path, f))]
        random.shuffle(images) # Shuffle images for random split

        # Determine target binary category
        target_category = ''
        if class_folder_name in damaged_classes:
            target_category = 'damaged'
        elif class_folder_name in undamaged_classes:
            target_category = 'undamaged'
        else:
            print(f"Warning: Class '{class_folder_name}' not assigned to 'damaged' or 'undamaged'. Skipping.")
            continue

        num_train = int(len(images) * train_split_ratio)
        train_images = images[:num_train]
        test_images = images[num_train:]

        # Copy images to the new train/test/category structure
        for img_name in train_images:
            src = os.path.join(original_class_path, img_name)
            dst = os.path.join(new_train_dir, target_category, img_name)
            shutil.copy(src, dst)

        for img_name in test_images:
            src = os.path.join(original_class_path, img_name)
            dst = os.path.join(new_test_dir, target_category, img_name)
            shutil.copy(src, dst)

        print(f"  Class '{class_folder_name}': {len(train_images)} train, {len(test_images)} test images moved to '{target_category}' category.")

print("Data splitting complete.")

# ============================================
# 3. DATASET PATHS (UPDATE IF NEEDED)
# ============================================
# Update train_dir and test_dir to the newly created split directories
train_dir = new_train_dir
test_dir = new_test_dir

print(f"Updated train_dir: {train_dir}")
print(f"Updated test_dir: {test_dir}")

# ============================================
# 4. IMAGE DATA GENERATOR + AUGMENTATION
# ============================================
img_height, img_width = 224, 224
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

print("Class Mapping:", train_gen.class_indices)

# ============================================
# 5. CNN MODEL
# ============================================
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.4),

    Dense(1, activation='sigmoid')
])

model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.0001),
    metrics=['accuracy']
)

model.summary()

# ============================================
# 6. TRAINING
# ============================================
history = model.fit(
    train_gen,
    epochs=20,
    validation_data=test_gen
)

# ============================================
# 7. SAVE THE MODEL
# ============================================
model.save("solar_panel_damage_model.h5")
print("Model saved successfully!")

# ============================================
# 8. ACCURACY & LOSS PLOTS
# ============================================
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Train", "Test"])

plt.subplot(1,2,2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["Train", "Test"])

plt.show()

# ============================================
# 9. CONFUSION MATRIX & CLASSIFICATION REPORT
# ============================================
y_pred = model.predict(test_gen)
y_pred = (y_pred > 0.5).astype(int)
y_true = test_gen.classes

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=["damaged", "undamaged"]))
```
